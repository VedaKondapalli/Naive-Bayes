```python
def remove_tags(string):
    removelist = ""
    result = re.sub('','',string)          
    result = re.sub('https://.*','',result)   
    result = re.sub(r'[^w'+removelist+']', ' ',result)    
    result = result.lower()
    return result
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
training_data['Review'] = training_data['Review'].apply(lambda x: ' '.join([words for words in str(x).split() if words not in (stop_words)]))
```

```python
import nltk
nltk.download('omw-1.4')
nltk.download('wordnet')
w_tokenizer = nltk.tokenize.WhitespaceTokenizer()
lemmatizer = nltk.stem.WordNetLemmatizer()
def lemmatize_text(text):
  st = ""
  for w in w_tokenizer.tokenize(text):
    st = st + lemmatizer.lemmatize(w) + " "
  return st
training_data['Review'] = training_data.Review.apply(lemmatize_text)
```

```python
from sklearn.preprocessing import LabelEncoder
New_Sentence = training_data['Review'].values
labels = training_data['Freshness'].values
encoder = LabelEncoder()
encoded_labels = encoder.fit_transform(labels)
```

```python
train_sentences, test_sentences, train_labels, test_labels = train_test_split(New_Sentence, encoded_labels, stratify = encoded_labels)
```

```python
from collections import defaultdict
training_data.reset_index(inplace = True)
```

```python
vec = CountVectorizer(max_features = 3000)
X = vec.fit_transform(train_sentences)
vocab = vec.get_feature_names_out()
X = X.toarray()
word_counts = {}
for l in range(2):
    word_counts[l] = defaultdict(lambda: 0)
for i in range(X.shape[0]):
    l = train_labels[i]
    for j in range(len(vocab)):
        word_counts[l][vocab[j]] += X[i][j]
word_counts

# Create word count dataframe
  #https://stackoverflow.com/questions/70640923/countvectorizer-object-has-no-attribute-get-feature-names-out
```

```python
def laplace_smoothing(n_label_items, vocab, word_counts, word, label):
    a = word_counts[label][word] + 0.0001
    b = n_label_items[label] + len(vocab)
    return math.log(a/b)
```

```python
def group_by_label(x, y, labels):
    data = {}
    for l in labels:
        data[l] = x[np.where(y == l)]
    return data
```

```python
def fit(x, y, labels):
    n_label_items = {}
    log_label_priors = {}
    n = len(x)
    grouped_data = group_by_label(x, y, labels)
    for l, data in grouped_data.items():
        n_label_items[l] = len(data)
        log_label_priors[l] = math.log(n_label_items[l] / n)
    return n_label_items, log_label_priors
```

```python
def predict(n_label_items, vocab, word_counts, log_label_priors, labels, x):
    result = []
    for text in x:
        label_scores = {l: log_label_priors[l] for l in labels}
        words = set(w_tokenizer.tokenize(text))
        for word in words:
            if word not in vocab: continue
            for l in labels:
                log_w_given_l = laplace_smoothing(n_label_items, vocab, word_counts, word, l)
                label_scores[l] += log_w_given_l
        result.append(max(label_scores, key=label_scores.get))
    return result
```

```python
labels = [0,1]
n_label_items, log_label_priors = fit(train_sentences,train_labels,labels)
pred = predict(n_label_items, vocab, word_counts, log_label_priors, labels, test_sentences)
print("Accuracy of prediction on test set : ", accuracy_score(test_labels,pred))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vect = TfidfVectorizer()
#fits the Tfidfvectorizer to your corpus, creating the vocabulary
X=vect.fit_transform(training_data['Review'])

#shows the fitted TfidfVectorizer and your corpus vocabulary
len(vect.vocabulary_)
vect.vocabulary_
```

```python
# Get the feature names (words) from the vocabulary
vocab = vect.get_feature_names_out()

# Create a dictionary to store the top 10 words for each class
top_words_by_class = {}

# Loop through each class
for label in labels:
    # Get the indices of data points belonging to the current class
    indices = np.where(encoded_labels == label)[0]
    # Calculate the sum of tf-idf scores for each word in the vocabulary for the current class
    word_scores = np.sum(X[indices], axis=0).A1
    # Sort the words based on their scores in descending order
    sorted_words_indices = np.argsort(word_scores)[::-1]
    # Get the top 10 words for the current class
    top_words = [vocab[i] for i in sorted_words_indices[:10]]
    # Add the top words to the dictionary
    top_words_by_class[label] = top_words

# Print the top 10 words for each class
for label, top_words in top_words_by_class.items():
    print("Class {}: {}".format(label, ', '.join(top_words)))
```
